## 45 - Triangular, Pentagonal, and Hexagonal
> Triangle, pentagonal, and hexagonal numbers are generated by the following formulae:
    <table><tr><td>Triangle</td>
    <td> </td>
    <td>$T_n=n(n+1)/2$</td>
    <td> </td>
    <td>$1, 3, 6, 10, 15, \dots$</td>
    </tr><tr><td>Pentagonal</td>
    <td> </td>
    <td>$P_n=n(3n - 1)/2$</td>
    <td> </td>
    <td>$1, 5, 12, 22, 35, \dots$</td>
    </tr><tr><td>Hexagonal</td>
    <td> </td>
    <td>$H_n=n(2n - 1)$</td>
    <td> </td>
    <td>$1, 6, 15, 28, 45, \dots$</td>
    </tr></table><p>It can be verified that $T_{285} = P_{165} = H_{143} = 40755$.</p>
    <p>Find the next triangle number that is also pentagonal and hexagonal.</p>



```python
limit = 10**5
found = False

while not(found):
    t = set(i*(i+1)//2 for i in range(286, limit))
    p = set(i*(3*i-1)//2 for i in range(166, limit))
    h = set(i*(2*i-1) for i in range(144, limit))

    for n in t:
        if n in p and n in h:
            print(n)
            found = True
            break

    limit *= 10
```
