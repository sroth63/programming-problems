## 44 - Pentagon Numbers
> Pentagonal numbers are generated by the formula, $P_n=n(3n-1)/2$. The first ten pentagonal numbers are: $$1, 5, 12, 22, 35, 51, 70, 92, 117, 145, \dots$$</p>
    <p>It can be seen that $P_4 + P_7 = 22 + 70 = 92 = P_8$. However, their difference, $70 - 22 = 48$, is not pentagonal.</p>
    <p>Find the pair of pentagonal numbers, $P_j$ and $P_k$, for which their sum and difference are pentagonal and $D = |P_k - P_j|$ is minimised; what is the value of $D$?</p>

Stricly speaking, my solution doesn't ensure $D$ is indeed minimal. However, looking at the forum on the Project Euler website, it seems that nobody has managed to write a Python code that ensures $D$ is minimal and that runs under a minute...

```python
dmax = 10000
pentagonals = set([(n*(3*n - 1))//2 for n in range(1,(5*dmax)//3)])

def search():
    d = 0
    while True:
        d += 1
        D = (d*(3*d - 1)) // 2
        for P in pentagonals:
            if 2*P + D in pentagonals and P + D in pentagonals:
                return D
            
print(search())
```
